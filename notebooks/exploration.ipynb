本 Notebook 用於初步查看 `sample_ocean_data.csv`，包含資料預覽、缺失值檢查、統計摘要與簡單圖表。

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
DATA_PATH = PROJECT_ROOT / 'data' / 'raw' / 'sample_ocean_data.csv'
```

## 讀取 sample 資料

```python
df = pd.read_csv(DATA_PATH)
df.head()
```

## 資料基本資訊

```python
print(df.info())
```

## 缺失值檢查

```python
df.isna().sum()
```

## 統計摘要

```python
df.describe()
```

## 簡單圖表

```python
df['date'] = pd.to_datetime(df['date'])

plt.figure(figsize=(10, 5))
for station, station_df in df.groupby('station'):
    station_df = station_df.sort_values('date')
    plt.plot(station_df['date'], station_df['sea_temperature_c'], marker='o', label=station)

plt.title('Sea Temperature Trend by Station')
plt.xlabel('Date')
plt.ylabel('Sea Temperature (C)')
plt.legend()
plt.grid(alpha=0.25)
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

## 初步觀察結論

- sample 資料包含五個測站，每個測站皆有多天觀測紀錄。
- 海溫在不同測站之間有差異，南部測站整體溫度較高。
- 折線圖可以快速看出不同測站隨日期變化的趨勢。
